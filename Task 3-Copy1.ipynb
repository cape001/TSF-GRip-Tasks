## Exploratory Data Analysis - Retail

## Load the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SampleSuperstore.csv')
```

```python
print(df.head())
```

```python
df.shape
```

```python
df = df.drop("Postal Code", axis = 1)
df.head()
```

```python
import seaborn as sns
sns.pairplot(df);
```

# **Let's begin with Ship Mode**
We evaluate our dataset on the basis of the shipmet mode

```python
set(df['Ship Mode'])
```

### We can see that we have 4 shipment modes

```python
plt.figure(figsize = (5,5))
textprops = {"fontsize":15}

plt.title('Shipment Modes')
plt.pie(df['Ship Mode'].value_counts(), labels = set(df['Ship Mode']), autopct = '%1.2f%%', textprops = textprops)

plt.show()
```

### What are the categories?

```python
plt.figure(figsize = (5,5))
plt.title('Categories')
textprops = {"fontsize":15}

plt.pie(df['Category'].value_counts(), labels = set(df['Category']), autopct = '%1.2f%%', textprops =textprops);
```

```python
plt.figure(figsize = (10,10))
df.groupby('Ship Mode')['Sales'].sum().nlargest(n = 100).plot(kind = 'bar', fontsize = 18 )
df.groupby('Ship Mode')['Profit'].sum().nlargest(n = 100).plot(kind = 'bar', fontsize = 18, color = 'green' )
plt.xlabel('Shipment Mode', fontsize = 18)
plt.ylabel('Sales/Profit', fontsize = 18)
plt.title("Top 100 Shipment Mode Sales vs Profit",fontsize = 20)
plt.style.use('seaborn')



plt.show()
```

## What we conclude from this is,  higehst profit and sales were made from standard class shipment

```python
plt.figure(figsize = (10,10))
df.groupby('Category')['Sales'].sum().nlargest(n = 100).plot(kind = 'bar', fontsize = 18 )
df.groupby('Category')['Profit'].sum().nlargest(n = 100).plot(kind = 'bar', fontsize = 18, color = 'green' )
plt.xlabel('Category', fontsize = 18)
plt.ylabel('Sales/Profit', fontsize = 18)
plt.title("Catefories vs Sales/Profit",fontsize = 20)
plt.style.use('seaborn')


plt.show()
```

# Technology and office supplies sales are high along with the profits. Even though the sales in furniture is high, the profit, not so much.

## Now lets talk abou the sub-categories

```python
plt.figure(figsize = (10,10))
plt.title('Sub-Categories')
plt.pie(df['Sub-Category'].value_counts(), labels = set(df['Sub-Category']), autopct = '%1.2f%%', textprops = {'fontsize':15});
```

```python
plt.figure(figsize = (10,10))
plt.style.use('seaborn')

df.groupby('Sub-Category')['Sales'].sum().nlargest(n = 100).plot(kind = 'bar', fontsize = 18 )
df.groupby('Sub-Category')['Profit'].sum().nlargest(n = 100).plot(kind = 'bar', fontsize = 18, color = 'green' )
plt.xlabel('Sub-Category', fontsize = 18)
plt.ylabel('Sales/Profit', fontsize = 18)
plt.title("Top 100 Shipment Mode Sales vs Profit",fontsize = 20)

plt.show()
```

## What we see from this is, the sales of Paper, Accessories, Binders and Chairs should be incresed as the profit will increase as well, whereas Tables and Bookcases are not doing well in the market.

```python
plt.figure(figsize = (10,10))
plt.style.use('seaborn')

df.groupby('Region')['Sales'].sum().nlargest(n = 100).plot(kind = 'bar', fontsize = 18 )
df.groupby('Region')['Profit'].sum().nlargest(n = 100).plot(kind = 'bar', fontsize = 18, color = 'green' )
plt.xlabel('Region', fontsize = 18)
plt.ylabel('Sales/Profit', fontsize = 18)
plt.title("Region vs Profit",fontsize = 20)

plt.show()
```

## The West and East regions are doing well, Central region needs a looking into.
